```python
import pandas as pd
```

```python
bori = pd.read_csv('DB/x_변수병합(주민등록세대수_인구수_주택유형).csv', encoding='CP949')
house = pd.read_csv("DB/가구원수별_수정.csv", encoding='CP949')
moving = pd.read_csv("DB/전입출_수정.csv", encoding='CP949') 
```

```python
print(bori.columns)
print(house.columns)
print(moving.columns)
```

```python
house_renamed = house.rename(columns={'시점': '연도'})
merge1 = pd.merge(bori, house_renamed, on=['행정구역', '연도'], how='inner')
merge1.head()
```

```python
moving_renamed = moving.rename(columns={'시점': '연도'})
merge2 = pd.merge(merge1, moving_renamed, on=['행정구역', '연도'], how='inner')
merge2.head()
```

```python
merge2.columns
```

```python
merged = merge2[['행정구역', '시도_x', '시군구_x', '연도', '주민등록세대수', '총인구수 (명)', '단독주택-계', '아파트',
       '연립주택', '다세대주택', '비주거용 건물 내 주택', '단독주택비율', '아파트비율', '연립주택비율', '다세대주택비율',
       '비주거용주택비율', '1인', '2인', '3인', '4인', '5인 이상', '총전입 (명)', '총전출 (명)', '순이동 (명)']]
```

```python
merged_renamed = merged.rename(columns={'시도_x' : '시도',
                                        '시군구_x' : '시군구',
                                        '총인구수 (명)': '총인구수',
                                        '1인':'1인가구',
                                        '2인':'2인가구',
                                        '3인':'3인가구',
                                        '4인':'4인가구',
                                        '5인 이상': '5인이상가구',
                                        '총전입 (명)':'총전입', 
                                        '총전출 (명)':'총전출', 
                                        '순이동 (명)':'순이동'
                                        })
merged_renamed.head()
```

```python
merged_renamed.info()
```

```python
print(set(bori['행정구역'].unique())-set(merged_renamed['행정구역'].unique()))
# 삭제된 행정구역 없음
```
